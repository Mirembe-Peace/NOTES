# **Lists 😎**
> Lists in python is used to store multiple items in a single variable and the values can change

**NOTES**
- lists in python can store items of different data types
- A list in python is opened with square brackets (like an array)

```python
countries = [ "United States", " India ", " China", "Brazil"]
countries
```

```python
countries[2]
```

#### **Operations that can be carried out on a list**
- Slicing - create sections of a list
- Append - adds an element an the end of the list
- Insert - adds an element to a specified position
- Join - concatinate or join 2 lists 
- Nesting - make a list part of another list
- remove - removes an element from the list by name
- pop - removes an element from the list by index (using the negative indices)
- del - function, removes an element from the list by index (using the positive indices)
- Sorting - sorting the elements of a list
- Update - updating an element in the list
- Copying - copying items of a list

```python

# Slicing
# This brings a section of the list 
# does not include the second value indicated

#list_name[ INDEX_OF_first_element: INDEX_OF_element before the last one to be included in the slice]

countries[1:3]
```

```python
# using the append method
countries.append("Newzealand")
countries
```

```python
# using the insert method
# list_name.insert(index_of_where_we_want_to_insert, element to insert)
countries.insert(3, "Spain")
countries
```

```python
countries.remove('Newzealand')
countries
```

```python
# joining two lists , we use the "+" operator .. we just add
countries_2 = ["Britain", "German", "Monacco"]
country = countries + countries_2
country
```

```python
# nested list
nested_list = [countries, countries_2]
nested_list
```

```python
# pop method
countries.pop(-6)
countries
```

```python
# del function
del countries[0]
countries
```

```python
# sorting a list
numbers = [3, 6, 8, 1, 34, 71, 16, 23, 89, 43]

numbers.sort()
numbers
```

```python
# reverse sorting (descending order)
numbers.sort(reverse = True)
numbers
```

```python
#updating an element in the list
numbers[2] = 33
numbers
```

```python
# copying 
# using the copy method
new_countries = countries.copy()
new_countries
```

```python
# using the slice technique
# not indicating the indices of the first and last element implies that we want the whole list
new_countries_2 = countries[:]
new_countries_2
```

# **Dictionaries 📖**

> A dictionary is an ordered list of elements containing a key and a value

**NOTES**
- A dictionary in python is opened with curly brackets 
- a key and a value together are called an ***item***

```python
dictionary_name = {
    'key_1': 'value_1',
    'key_2': 'value_2'
    }
```

```python
# creating a dictionary
hotspot_data = {
    'position' : 'center',
    'size' : 45,
    'text' : 'first'
}

hotspot_data
```

```python
# retrieving only the keys
# using the .keys() method

hotspot_data.keys()
```

```python
# retrieving the values oonly using .values() method
hotspot_data.values()
```

```python
# retrieving items using the .items() method
hotspot_data.items()
```

#### **Operations that can be carried out on a dictionary**
- Adding an item
- Updating a value of a key
- Copying a dictionary 
- removing a key from a dictionary

**NB:** Changes in the original dictionary do not affect its copy after the copy has been created 

```python
# adding an item to a dictionary
# dictionary_name['key'] = value
hotspot_data['audio_file'] = 3
hotspot_data
```

```python
# updating a value using the .update() method 
# dictionary_name.update({'key' : 'value'})
hotspot_data.update({'audio_file' : 7})
hotspot_data
```

```python
# copying a dictionary using the .copy() method
new_dict = hotspot_data.copy()
new_dict
```

```python
# removing an element
# using the .pop()
# dictionary_name.pop("key")
hotspot_data.pop('height')
hotspot_data
```

```python
#using the def technique
# del dictionary_name['key']
del hotspot_data['audio_file']
hotspot_data
```

```python
# removing all elements in a dictionary using the .clear() method
new_dict.clear()
new_dict
```

# **Loops ➿**


1. **For Loops**

**NOTE:** an iterable element can be a list, dictionary, array etc.. 

```python
# for <create_variable> in <iterable_object(variable)>:
#     <code>
for country in countries:
    print(country)
```

```python
# using enumerate(iterable_object) function to count the number of loops
# for number_of_iteration, created-varaible in iterable_object.enumerate():
for i, country_tag in enumerate(countries):
    print(country_tag, i)
    
```

```python
# for loop on a dictionary
# for key, value in dictionary_name.items():
#     <code>

for key, value in hotspot_data.items():
    print(key, value)
```
